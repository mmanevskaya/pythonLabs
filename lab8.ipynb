```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Загрузить файл titanic.csv в с помощью pandas
df = pd.read_csv("./titanic.csv")
```

```python
# продемонстрировать загруженный датафрейм
df
```

```python
# Только мужчины от 30 до 40 лет
df[(df["Sex"] == "male") & (df["Age"] > 30) & (df["Age"] < 40)]
```

```python
# Только выжившие пассажиры 1 класса или женщины моложе 25 лет
df[((df["Survived"] == 1) & (df["Pclass"] == 1)) | ((df["Sex"] == "female") & (df["Age"] < 25))]
```

```python
# Построить график (сгруппировать с помощью функции groupby) 
# распределения по полу и по классу   groupby(["Sex","Pclass"])
plt.rcParams["figure.figsize"] = (14, 7)
pd.crosstab(df['Sex'],df['Pclass']).plot.bar()
plt.xticks(rotation=45)
plt.show()
```

```python
# Сделать срез пассажиров только моложе 25 лет.
# По полученному срезу построить график, сгруппировав по 
# полу, классу и выживанию groupby(["Sex","Pclass","Survived"])
plt.rcParams["figure.figsize"] = (14, 7)
df[df["Age"] < 25].groupby(["Sex","Survived", "Pclass"])[["Sex","Survived", "Pclass"]].size().unstack(fill_value=0).plot.bar()
plt.xticks(rotation=45)
plt.show()
```
