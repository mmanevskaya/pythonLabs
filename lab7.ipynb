```python
import numpy as np
from numpy import sin, pi
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

---
$$\normalsize\text{Задание №1}$$
Отобразить график: $$y = 2*\sin(pi*3x)$$

```python
x = np.linspace(1, 20, 100)
```

```python
y = 2 * sin(pi * 3 * x)
```

```python
plt.figure(figsize=(14,8))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

$$\normalsize\text{Задание №2}$$

Построить диаграмму пирога (pie), на основе следующих данных:
Калорийность обеда:
- Винегрет овощной 100ккал
- Суп 150ккал
- Котлеты 200ккал
- Гречка 140ккал
- Чай 20ккал
- Тортик 450ккал

```python
kcal = [100, 150, 200, 140, 20, 450]
dish = ['Винегрет овощной', 'Суп', 'Котлеты', 'Гречка', 'Чай', 'Тортик']

plt.figure(figsize=(14,8)) 
plt.title("Калорийность обеда в % (ккал)")
plt.pie(x = kcal,
        labels = dish,
        colors = sns.color_palette('pastel')[0:6], 
        autopct='%.0f%%')

plt.show()
```

$$\normalsize\text{Задание №3}$$

Построить диаграмму столбцов (bar) на основе следующих данных:
Количество осадков в месяце:
+ Январь - 39.4
+ Февраль - 35.5
+ Март - 40.1
+ Апрель - 48.4
+ Май - 79.8
+ Июнь - 65.9
+ Июль - 74.1
+ Август - 43.9
+ Сентябрь 49.2
+ Октябрь 40.3
+ Ноябрь 39.4
+ Декабрь 43.2

```python
Month = ['Январь',
        'Февраль',
        'Март',
        'Апрель',
        'Май',
        'Июнь',
        'Июль',
        'Август',
        'Сентябрь',
        'Октябрь',
        'Ноябрь',
        'Декабрь']

Count = [39.4,
         35.5,
         40.1,
         48.4,
         79.8,
         65.9,
         74.1,
         43.9,
         49.2,
         40.3,
         39.4,
         43.2]
```

```python
plt.figure(figsize=(14,8))
plt.title("Столбчатая диаграмма на основе данных о количестве осадков за месяц:")

sns.barplot(x=Month, y=Count)

plt.xlabel("Месяц")
plt.ylabel("Количество осадков")
plt.show()
```
